```python
import pandas as pd
```

```python
df = pd.read_csv('BBUWVE.csv')
```

```python
df
```

```python
df['Total'] = 100
```

```python
df['Obtain_Marks'] = (df.quiz_1+df.quiz_2+df.quiz_3+df.final_exam)
#df['Obtain_Marks'] = df.iloc[:,2:].sum(axis=1)
```

```python
df["Percentage"] = (df.Obtain_Marks / df.Total) * 100
```

```python
def grade(x):
    if x >= 80:
        return 'A'
    if x >= 70:
        return 'B'
    if x >= 60:
        return 'C'
    if x < 60:
        return 'Fail'
```

```python
df["Grade"] = df['Percentage'].apply(grade)
```

```python
df['Classes'] = df['roll_no'].str.split('/').str[1]
```

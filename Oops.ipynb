```python
class book:
    def __init__(self,title,author,publishing_date,page_numbers,chapters,isbn):
        self.title = title
        self.author = author
        self.publishing_date = publishing_date
        self.page_numbers = page_numbers
        self.chapters = chapters
        self.isbn = isbn
    def gettitle(self):
        return self.title
    def settitle(self,new_title):
        self.title = new_title
    def getauthor(self):
        return self.author
    def setauthor(self,new_author):
        self.author = new_author
    def getpublishing_date(self):
        return self.publishing_date
    def setpublishing_date(self,new_publishing_date):
        self.publishing_date = new_publishing_date
    def getchapters(self):
        return self.chapters
    def setchapters(self,new_chapters):  
        self.chapters = new_chapters
    def getisbn(self):
        return self.isbn
    def setisbn(self,new_isbn):    
        self.isbn = new_isbn  
```

```python
customers = pd.read_csv("customers.csv")
products = pd.read_csv("products.csv")
orders = pd.read_csv("orders.csv")
```

```python
half = orders.merge(products , on = "product_id")
```

```python
full = half.merge(customers , on = "customer_id")
```

```python
full.order_date = pd.to_datetime(full.order_date)
```

# Repeating Customer

```python
customer_count_2022 = full[full['order_date'].dt.year == 2022]['customer_id'].value_counts()
customer_count_2023 = full[full['order_date'].dt.year == 2023]['customer_id'].value_counts()
customer_count_2024 = full[full['order_date'].dt.year == 2024]['customer_id'].value_counts()
```

```python
Repeating_customer_2022 = customer_count_2022[customer_count_2022 > 1]
Repeating_customer_2023 = customer_count_2023[customer_count_2023 > 1]
Repeating_customer_2024 = customer_count_2024[customer_count_2024 > 1]
```

```python
print(f"Total no of repeating customers of 2022 are: {(len(Repeating_customer_2022))} out of {(len(customer_count_2022))}")
print(f"Total no of repeating customers of 2023 are: {(len(Repeating_customer_2023))} out of {(len(customer_count_2023))}")
print(f"Total no of repeating customers of 2024 are: {(len(Repeating_customer_2024))} out of {(len(customer_count_2024))}")
```

# Max Selling product
    According to the repeated customer

```python
products_counts_2022 = full[full['order_date'].dt.year == 2022]['product_id'].value_counts()
products_counts_2023 = full[full['order_date'].dt.year == 2023]['product_id'].value_counts()
products_counts_2024 = full[full['order_date'].dt.year == 2024]['product_id'].value_counts()
```

```python
most_repeated_2022 = products_counts_2022.idxmax()
most_repeated_2023 = products_counts_2023.idxmax()
most_repeated_2024 = products_counts_2024.idxmax()
```

```python
most_selling_2022 = full[full['product_id']==most_repeated_2022]['product_name'].iloc[0]
most_selling_2023 = full[full['product_id']==most_repeated_2023]['product_name'].iloc[0]
most_selling_2024 = full[full['product_id']==most_repeated_2024]['product_name'].iloc[0]
```

```python
print(f"Most selling product of year 2022 is: {(most_selling_2022)}\nThe no of sells are: {(len(most_selling_2022))}")
print('-----------------------------------------------')
print(f"Most selling product of year 2023 is: {(most_selling_2023)}\nThe no of sells are: {(len(most_selling_2023))}")
print('-----------------------------------------------')
print(f"Most selling product of year 2024 is: {(most_selling_2024)}\nThe no of sells are: {(len(most_selling_2024))}")
```

# Product To Discontinue

```python
least_repeated_2022 = products_counts_2022.idxmin()
least_repeated_2023 = products_counts_2023.idxmin()
least_repeated_2024 = products_counts_2024.idxmin()
```

```python
least_selling_2022 = full[full['product_id']==least_repeated_2022]['product_name'].iloc[0]
least_selling_2023 = full[full['product_id']==least_repeated_2023]['product_name'].iloc[0]
least_selling_2024 = full[full['product_id']==least_repeated_2024]['product_name'].iloc[0]
```

```python
print(f"Most selling product of year 2022 is: {(least_selling_2022)}\nThis product needs to be discontinued")
print('-----------------------------------------------')
print(f"Most selling product of year 2023 is: {(least_selling_2023)}\nThis product needs to be discontinued")
print('-----------------------------------------------')
print(f"Most selling product of year 2024 is: {(least_selling_2024)}\nThis product needs to be discontinued")
```

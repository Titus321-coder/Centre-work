```python
class Car:
    def __init__(self,car_name,car_model,price,color):
        self.car_name = car_name
        self.car_model = car_model
        self.price = price
        self.color = color
    def __repr__(self):
        return self.__str__()
    def __str__(self):
        return f"Available Cars:\n{self.car_name}\nModel:{self.car_model} is ${self.price} in {self.color}"
class showroom:
    def __init__(self):
        self.purchased_cars = []
    def add_car(self,car):
        self.purchased_cars.append(car) 
    def all_cars(self):
        for car in self.purchased_cars:
            print(f"Model:{self.car_model} is ${self.price} in {self.color}")
        
```

```python
Shop = showroom()
```

```python
class Student:
    def __init__(self,name,id):
        self.name = name
        self.id = id
        self.add_student = []
    def register_student(self,std):
        with open('Register.txt','a') as file:
            info = self.add_student()
            file.write(info)
    def show_all(self):
        with open('Register.txt','r') as show:
            data = show.readlines()
            for std in data:
                return std
```

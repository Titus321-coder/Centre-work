```python
def greet(func):
    def wrapper(*args,**kwargs):
        print('Welcome')
        myfunc = func(*args,**kwargs)
        print('Good Bye')
        return myfunc
    return wrapper    

def confirmation(func):
    def wrapper(*args,**kwargs):
        user_confirm = input("Enter Y to confirm OR N to cancel...")
        if user_confirm.lower() == 'Y':
            func(*args,**kwargs)
        return wrapper    
```

```python
class bank:
    def __init__(self,Account_Title,Balance):
        self.Account_Title = Account_Title
        self.Balance = Balance
    def __repr__(self):
        return self.__str__()
    @greet    
    @confirmation
    def __str__(self):
        return f"Your Title is: {self.Account_Title}, Your remaining Balance is: {self.Balance}"
    @confirmation    
    def __add__(self,other):
        self.balance += other
    @confirmation    
    def __sub__(self,other):
        self.balance -= other    
```
